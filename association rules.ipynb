```python
import pandas as pd

data=pd.read_excel('Items Complete.xls')
data
```

```python
data=data.drop('ID',axis=1)
data=data.drop('ostancod',axis=1)
data.head()
```

```python
from mlxtend.frequent_patterns import apriori


fre_item=apriori(data,min_support=0.4, use_colnames=True)
fre_item
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(fre_item,metric='confidence',min_threshold=0.4)

rules=association_rules(fre_item,metric='lift',min_threshold=1)
rules
```

```python
rules['len']=rules['antecedents'].apply(lambda x: len(x))

rules
```

```python
rules[(rules['len']>=2) &
      (rules['confidence']>0.7) &
      (rules['lift']> 1.03)
           ]
```
