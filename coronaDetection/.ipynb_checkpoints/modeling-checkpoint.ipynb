```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('coronaClean.xlsx')
df
```

```python
del df['Unnamed: 0']
```

```python
#remove irrevalant features with RF exe

li = ['Previous_history_of_corona','Immunodeficiency',
      'Inflammation_skin lesion','Limb plegia']

df_c = df.drop(li,axis=1)
```

### Ø¨Ø¯ÙˆÙ† Ø§Ù†ØªØ®Ø§Ø¨ ÙˆÛŒÚ˜Ú¯ÛŒ

```python
X=df.drop('Assignment_status',axis=1)

Y= df['Assignment_status']
```

### Ø¨Ø§ Ø§Ù†ØªØ®Ø§Ø¨ ÙˆÛŒÚ˜Ú¯ÛŒ#

```python
X=df_c.drop('Assignment_status',axis=1)

Y= df_c['Assignment_status']
```

### ØªÙ‚Ø³ÛŒÙ… Ø¯Ø§Ø¯Ù‡ Ù‡Ø§ Ø¨Ù‡ Ø¢Ù…ÙˆØ²Ø´ÛŒ Ùˆ ØªØ³Øª

```python

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,
                                               random_state=42,shuffle=True)
```

```python
x_train.shape
```

```python

#layers for fully connected
from keras.models import Sequential
from keras.layers import Dense

classifier=Sequential()

#first layer in fc(MLP with two layers - 9 attributes)


classifier.add(Dense(17, kernel_initializer='uniform', activation='relu' , input_dim=37))

#second layer in fc(MLP with two layers)
classifier.add(Dense(17, kernel_initializer='uniform', activation='relu'))


#add output layer
classifier.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

#compile(optimizing setting)
classifier.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])

classifier.fit(x_train,y_train, batch_size=10, epochs=3)

y_pred=classifier.predict(x_test)
```
