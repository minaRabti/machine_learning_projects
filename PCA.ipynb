```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition



def column(matrix,i):
    return[row[i] for row in matrix]

#load data
iris=pd.read_excel('dataset.xls')
x=iris[['Sepal Width','Sepal Length','Peta Width','Peta Length']]
y=iris['Types']

label=y
#PCA
pca3D=decomposition.PCA(n_components=3)
x1=pca3D.fit(x)
x1=pca3D.transform(x)

#figure
col1=column( x1 ,0)
col2=column( x1 ,1)
col3=column( x1 ,2)
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(col1,col2,col3)
plt.show()
```
