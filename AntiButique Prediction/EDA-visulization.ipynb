```python
data2 = pd.read_excel('sheet2.xlsx')
data2.head()
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = "Province", y = "simple_hospital", data = data2)
plt.xticks(rotation=40)
plt.xlabel('Province')
plt.ylabel('Number of sample hospitals')
plt.title('Show the number of sample hospitals in each province' , fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(10,4))

labels=['hos_2011', 'hos_2012', 'hos_2013' , 'hos_2014' , 'hos_2015']

colors=['blue','red','yellow','green','brown']

explode=[0,0,0,0,0.1]

values=[data2.hos_2011.mean(),
        data2.hos_2012.mean(),
        data2.hos_2013.mean(),
        data2.hos_2014.mean(),data2.hos_2015.mean()]


plt.pie(values,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True)
plt.legend(['hos_2011','hos_2012','hos_2013','hos_2014','hos_2015'] , loc=5)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
data2.hist(figsize=(12,12),layout=(4,4))
plt.show()
```

```python
sns.kdeplot(data2['simple_hospital'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('simple_hospital Kde Plot System Analysis')
plt.show()
```

```python
sns.kdeplot(data2['hos_2011'],shade=True,color='red')
sns.kdeplot(data2['hos_2012'],shade=True,color='blue')
sns.kdeplot(data2['hos_2013'],shade=True,color='green')
sns.kdeplot(data2['hos_2014'],shade=True,color='yellow')
sns.kdeplot(data2['hos_2015'],shade=True,color='cyan')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('hos_2011 vs hos_2015 Kde Plot System Analysis')
plt.show()
```
