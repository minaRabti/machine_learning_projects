```python
#حذف ویژگی زیر به دلیل داده های از دست رفته بالا تا 90 درصد

data=data.drop('سی تی اسکن',axis=1)
```

```python
data.columns=['Age','Sex','coronaTest','Previous_history_of_corona','Then',
             'Cough','Muscular_pain','respiratory_distress','Loss_of_consciousness',
             'Olfactory','Sense_of_taste','convulsions','Stomach ache','nausea','Vomit',
             'diarrhea','Anorexia','Headache','dizziness','paresis_of_organs','Limb plegia',
             'Chest_pain','Inflammation_skin lesion','DaysOf','Intubation','Po2 status',
              'CT_Scan','cancer','Chronic_liver_diseases','diabetes','Chronic blood diseases',
             'Immunodeficiency','pregnancy','Heart_disease','Chronic_kidney_diseases',
              'Asthma','History_of_hypertension','Assignment_status']

```

```python
data.isnull().sum()
```

سطرهایی با 5 سطر به پایین مقدار از دست رفته را حذف می کنیم

```python
liRemove = ['Then','Cough','Muscular_pain','respiratory_distress','Loss_of_consciousness']


data=data.dropna(subset=liRemove)
data.shape
```

```python
from sklearn.impute import SimpleImputer


sm=SimpleImputer(strategy='most_frequent')

mat=sm.fit_transform(data)

data_c=pd.DataFrame(mat,columns=data.columns)
```

```python

data_c.info()
```

```python
data_c.drop_duplicates(inplace=True)
```

```python
data_c.shape
```

```python
174510 - 170288
```

```python
data_c['Age']=data_c['Age'].astype(float)
```

```python
data_c['Age'].describe()
```

```python
data_c['Sex']=data_c['Sex'].replace(['مرد'],'1')

data_c['Sex']=data_c['Sex'].replace(['زن'],'0')
```

```python
data_c=data_c.replace(['دارد'],'1')

data_c = data_c.replace(['ندارد'],'0')
```

```python
data_c
```

```python
data_c=data_c.replace(['بله'],'1')

data_c = data_c.replace(['خیر'],'0')
```

```python
data_c
```

```python
data_c=data_c.replace(['بلی'],'1')
data_c.head()
```

```python
data_c=data_c.replace(['مثبت'],'1')

data_c=data_c.replace(['منفی'],'0')
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()




data_c['History_of_hypertension']=le.fit_transform(data_c['History_of_hypertension'])
```
