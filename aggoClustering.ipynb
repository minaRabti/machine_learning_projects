```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Mall_Customers.csv')
data
```

```python
data.set_index('CustomerID',inplace=True)
```

```python
x=data.iloc[:,[2,3]]
```

```python
import scipy.cluster.hierarchy as sch

ch1=sch.linkage(x, method='single')

import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))

plt.title('h-clustering')

dend=sch.dendrogram(ch1)

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

agClustering =AgglomerativeClustering(n_clusters=6,linkage='ward')

agClustering.fit(data)
```

```python
agClustering.labels_
```

```python
x=data.values
```

```python
plt.scatter(x[:,2],x[:,3], c=agClustering.labels_,s=50)

plt.xlabel('income')
plt.ylabel('score')

plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(data,agClustering.labels_,metric='euclidean')
```
