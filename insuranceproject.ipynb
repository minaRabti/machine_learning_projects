```python
Required_Data = data[['بيمه گذار', 'تاریخ شروع', 'حق بيمه کل(با ماليات)']]
Required_Data
```

```python
Required_Data.columns = ['CustomerID','IssueDate','TotalPremium']
Required_Data
```

```python
Required_Data.info()
```

```python
Required_Data['IssueDate'] = pd.to_datetime(Required_Data['IssueDate'],
                                            format = '%Y-%m-%d %H:%M:%S',errors = 'coerce')
Required_Data
```

```python
Required_Data.info()
```

```python
Required_Data.isnull().sum()
```

```python
Required_Data=Required_Data.dropna(subset=['IssueDate'])
```

```python
Required_Data.info()
```

```python
Required_Data['IssueDate'].min(),Required_Data['IssueDate'].max()
```

```python
present = dt.datetime.today()
present
```

```python
RFM= Required_Data.groupby('CustomerID').agg({'IssueDate': lambda date: (present - date.max()).days,
                                        'CustomerID': lambda num: len(num),
                                        'TotalPremium': lambda price: price.sum()})
```

```python
RFM
```

```python
RFM.columns = ['Recency','frequently','Monetary']
RFM
```

```python
RFM['frequently'].value_counts()
```

```python
RFM.info()
```

```python
from sklearn.cluster import KMeans

WCSS = []
for i in range (2,11):
    KM = KMeans(n_clusters=i, max_iter =900 ,random_state=101)
    KM.fit(RFM)
    WCSS.append(KM.inertia_)
```

```python
plt.figure(figsize=(7,7))
plt.plot(range(2, 11), WCSS)

plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS") 

plt.show()
```

```python
kmeans = KMeans(n_clusters = 5,  max_iter =900 ,random_state=101)
y_kmeans = kmeans.fit_predict(RFM)

cluster_labels = kmeans.labels_ 
final_data = RFM.assign(Cluster = cluster_labels) 
final_data
```
