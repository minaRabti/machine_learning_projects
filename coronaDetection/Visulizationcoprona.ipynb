```python

ex=[0,0,0,0,0]
plt.figure(figsize=(7,7))

plt.pie(df['History_of_hypertension'].value_counts().values
        ,labels= df['History_of_hypertension'].value_counts().index,autopct='%1.1f%%',
        explode=ex, shadow=True
       )

plt.title('History_of_hypertension',color='black',fontsize=10)
plt.show()
```

```python

ex=[0,0,0.1]
plt.figure(figsize=(5,5))

plt.pie(df['coronaTest'].value_counts().values
        ,labels= df['coronaTest'].value_counts().index,autopct='%1.1f%%',
        explode=ex, shadow=True
       )

plt.title('coronaTest',color='black',fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(5,5))

sns.barplot(x = df['Assignment_status'].value_counts().index ,
            y= df['Assignment_status'].value_counts().values)

plt.xlabel('Assignment_status')
plt.ylabel('Frequency')

plt.title('Show of Assignment_status of corona(died/live)')
plt.show()
```
