```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Mall_Customers.csv')
data
```

```python
data.info()
```

```python
data.set_index('CustomerID',inplace=True)
```

```python
data
```

```python
#the elbow method

from sklearn.cluster import KMeans

SSE = []

for i in range(2,11):
    
    km=KMeans(n_clusters=i,max_iter=200,random_state=42)
    
    km.fit(data)
    
    SSE.append(km.inertia_)
    
    
SSE    
    
```

```python
plt.plot(range(2,11),SSE)

plt.xlabel('No of Clusters')

plt.ylabel('SSE')

plt.show()
```

Optimized No cluster=6

```python
kmeans = KMeans(n_clusters=6,max_iter=200,random_state=42)

kmeans.fit(data)

```

```python
cluster=kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
dataCluster=data.assign(cluterlabel=cluster)

dataCluster
```

```python
t=dataCluster[dataCluster['cluterlabel']==5]
```
